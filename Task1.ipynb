```python
'''Task1
data analysis using Python'''
#Data Loading
import pandas as pd
df=pd.read_csv("student-mat.csv",sep=';')#Loading data from the csv file to data frame
print('First Few rows are:\n',df.head())#it will return by default first 5 rows


```

```python
#Data Exploration
print("Dataset info")
print(df.info())
```

```python
#checking for missing values
print(df.isnull().sum())
```

```python
#return the data type of each column
result=df.dtypes
print(result)
```

```python
#.shape
#Understanding the dataset size
print(df.shape)
```

```python
#Data Cleaning
#remove duplicates
df=df.drop_duplicates()
#Data Analysis
#Question 1: What is the average score in Math(G3)?
averagescore=df['G3'].mean()
print(f"The average score is {averagescore:.2f}")
```

```python
#Question 2: How many students scored above 15 in their final grade?
students_above15_count=len(df[df['G3']>15])
print(f"No of students who scored above 15 is {students_above15_count}")
```

```python
#Question 3:Is there a correlation between study time and final grade(G3)?
relation=df['studytime'].corr(df['G3'])
print(f'The correlation between study time and final grade is {relation:.2f}')
```

```python
#Question 4:which gender has higher average final grade(G3)?
average_grade_by_gender=df.groupby('sex')['G3'].mean()
print(f'Average Final Grade BY Gender:\n {average_grade_by_gender}')
```

```python
#Data visualization
#Histogram of final grades
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.hist(df['G3'],bins=10,color='red',edgecolor='black')
plt.title('Histogram of final grades')
plt.xlabel('Grades')
plt.ylabel('no of students')
plt.show()
```

```python

#create a scatter plot of study time and final grades(G3)
import seaborn as sns
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='studytime',y='G3',hue='sex')
plt.title("Study time VS grades")
plt.xlabel('Study time')
plt.ylabel('Final grades')
plt.legend(title='Gender')
plt.show()

```

```python
#create a bar chart comparing the average scores of male and female students
average_grade_by_gender.plot(kind='bar',color=['blue','orange'])
plt.title("average final grade by gender")
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.show()
```
